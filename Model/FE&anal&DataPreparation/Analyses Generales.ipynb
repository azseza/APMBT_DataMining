```python
import numpy as np 
import seaborn
d
```

```python
#un petit traitement sur le nooms de colmones pour plus de lisibilité
d= d.rename(columns={"Open": "Open_TunIndex", "ouverture": "Open_Delice", "ouverture.2":"ouverture_sotipapier",
                   "ouverture.3":"ouverture_mbps","ouverture.4":"ouverture_sah","ouverture.5":"ouverture_cellcmod",
                   "ouverture.6":"ouverture_euro_cycles","ouverture.7":"ouverture_cityCars","ouverture.8":"ouverture_telnet",
                   "ouverture.9":"open_Telnet2","ouverture.10":"open_Ennakl","ouverture.11":"open_TunisRe",
                   "ouverture.12":"open_hannibalLease","ouverture.13":"open_Artes","ouverture.14":"open_Sopat",
                   "ouverture.15":"open_Tpr","ouverture.16":"open_Adwya","ouverture.17":"open_Essoukna",
                   "ouverture.18":"Wifak","ouverture.42":"open_Sfbt","ouverture.41":"open_TunisAir",
                   "ouverture.40":"open_tijari","ouverture.39":"open_biat","ouverture.38":"open_BH",
                   "ouverture.37":"open_Tl","ouverture.36":"open_Bt","ouverture.35":"open_STB",
                   "ouverture.25":"open_Sotetel","ouverture.26":"open_Star","ouverture.27":"open_Pgh",
                   "ouverture.28":"open_Atl","ouverture.29":"open_Cil","ouverture.30":"open_Sipmar"
                    }, index={"Unnamed: 0.1" : "Date"})
for i in list(d.columns):
    print(i)
```

```python
#on n'appliquera cette analyse que sur les prix d'ouverturee des diffrentes valuers ainsi que sur les variables 
#socio économique 
#(TunIndex, SFBT, DHL, Attijari, BH, TelNet... Import/export, IPI, MTT ... )
anal_corr = d[['IPI', 'MTT', 'BalCom', 
               'Open_TunIndex', 'Import', 'Export', 
               'Pib1', 'Open_Delice', 'ouverture_sotipapier', 
               'ouverture_mbps', 'ouverture_sah', 'ouverture_cellcmod', 
               'ouverture_cityCars', 'ouverture_telnet', 'open_Telnet2', 
               'open_Ennakl', 'open_TunisRe', 'open_hannibalLease', 
               'open_Artes', 'open_Sopat', 'open_Tpr', 
               'open_Adwya', 'open_Essoukna', 'Wifak', 
               'open_Sotetel', 'open_Star', 'open_Pgh', 
               'open_Atl', 'open_Cil', 'open_Sipmar', 
               'open_STB', 'open_Bt', 'open_Tl', 
               'open_BH', 'open_biat', 'open_tijari', 
               'open_TunisAir', 'open_Sfbt', 
              ]]
```

```python
anal_corr = anal_corr.corr(method='pearson')
anal_corr.reset_index()
#del anal_corr.index.name
anal_corr.head(10)
```

```python
mask = np.zeros_like(anal_corr)
mask[np.triu_indices_from(mask)] = True
plt.figure(figsize=(15,8))
seaborn.heatmap(anal_corr, cmap='RdYlGn', vmax=1.0, vmin=-1.0 , mask = mask, linewidths=0.01, cbar=True)
plt.yticks(rotation=0) 
plt.xticks(rotation=90) 
plt.show()
plt.savefig('corr.png')
```
