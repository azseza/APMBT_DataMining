```python
def realDate(fDate):
    mont = fDate[:3]
    day = fDate[4:6]
    if len(day) == 1 :
        day = '0'+day
    year = fDate[-4:]
    monts = dict
    monts ={'Jan': '01', 'Sep': '09', 'Dec': '12', 'May': '05',
        'Feb': '02','Jul': '07','Aug': '08','Oct': '10','Nov': '11','Apr': '04','Jun': '06','Mar': '03'}
    nmont = str(monts[mont])
    rDate = year+'-'+nmont+'-'+day
    return rDate
```

```python
for i in range(770):
    tunInx.at[i,"Date"] = realDate(tunInx.at[i,"Date"])
    
```

<h3>2/- Suprresion des collnes en NaN : </h3>

```python
a = pd.read_csv('test5.csv')
```

```python
a = a[a['Vol.'].notna()]
a
```

```python
#replacing missing values
a.fillna(method='ffill', inplace=True)
a.fillna(method='bfill', inplace=True)
a
```

```python
import numpy as np 
for k in a.columns : 
    print(k)
```

```python
a.set_index("Unnamed: 0")
```

```python
#dealing with the Vol. column 
for i in range(1100):
    try :
        b = a.at[i,"Vol."]
        if b[-1:] == 'K':
            a.at[i,"Vol."] = float(b[:-1])*1000
        elif b[-1:] == 'M':
            a.at[i,"Vol."] = float(b[:-1])*1000000
    except KeyError : 
        print('no row for index # '+ str(i))
    except TypeError : 
        print('index #  '+str(i)+' allready treted, moving on .. ')
```

```python
a['Vol.']
```

```python
a.dropna(how='all', axis=1)
```

```python
#handeling the '% in the ref raw '
for i in range(1076):
    try:
        a.at[i,"Change %"] = a.at[i,"Change %"][:-1] 
    except KeyError : 
        print("jkabf")
```

```python
a.set_index('Unnamed: 0')
```

```python
a.to_csv('Data.csv') #saving raw clean Data
```
